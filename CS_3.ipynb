```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```
imagem = cv2.imread('buraco_01.jpg')
```

```
# Convertendo a imagem para escala de cinza
gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Aplicando um filtro de suavização
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Segmentando a imagem
thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 2)

# Encontrando os contornos na imagem já segmentada
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Definindo a area mínima e máxima
area_min = 1000
area_max = 5000

# Inicializando o contador de buracos da imagem
num_buracos = 0

# Percorrendo os contornos e analisando suas propriedades
for contour in contours:
    # Calculando a área do contorno
    area = cv2.contourArea(contour)

    # Aplicando o filtro de área
    if area_min < area < area_max:
        # Contando os buracos da imagem
        num_buracos += 1

        # Desenhando um retângulo ao redor do buraco da imagem
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Exibindo a quantidade de buracos encontrados
print("Número de buracos: ", num_buracos)

# Exibindo a imagem com os buracos em destaque
cv2_imshow(imagem)
```

```
#Utilizando a biblioteca IPython e a funcionalidade display para possibilitar o usuário de carregar a imagem que desejar

import ipywidgets as widgets
from IPython.display import display

upload = widgets.FileUpload(accept='.jpg,.png', multiple=False)

display(upload)
```

```
image_data = upload.data[0]
```

```
#Salvando a imagem que foi carregada pelo usuário
from PIL import Image
import io

# Carregar a imagem a partir dos dados
imagem_processada = Image.open(io.BytesIO(image_data))

display(imagem_processada)
imagem_processada.save('imagem_processada.jpg')
```
